```python
import pandas as pd

df = pd.read_csv("../data/raw/occurrences.csv")

print(df.shape)
```

```python
cols = [
    "accepted_name",
    "early_interval",
    "late_interval",
    "collection_no",
    "environment",
    "formation",
    "lat",
    "lng",
    "max_ma",
    "min_ma"
]

clean = df[cols].copy()

clean.dropna(
    subset=[
        "accepted_name",
        "early_interval",
        "collection_no"
    ],
    inplace=True
)

print(clean.shape)
```

```python
valid_intervals = [
    "Changhsingian",
    "Griesbachian",
    "Dienerian",
    "Smithian",
    "Spathian",
    "Aegean",
    "Bithynian",
    "Pelsonian",
    "Illyrian",
    "Anisian"
]

clean = clean[
    clean["early_interval"].isin(valid_intervals)
].copy()

print(clean.shape)
```

```python
clean["early_interval"].value_counts()
```

```python
interval_order = [
    "Changhsingian",
    "Griesbachian",
    "Dienerian",
    "Smithian",
    "Spathian",
    "Aegean",
    "Bithynian",
    "Pelsonian",
    "Illyrian",
    "Anisian"
]

richness = (
    clean.groupby("early_interval")
    ["accepted_name"]
    .nunique()
    .reindex(interval_order)
)

print(richness)
```

```python
collections = (
    clean.groupby("early_interval")
    ["collection_no"]
    .nunique()
    .reindex(interval_order)
)

print(collections)
```

```python
comparison = pd.DataFrame({
    "richness": richness,
    "collections": collections
})

print(comparison.corr())
```
